## Data Cleaning with pandas


*   Empty cell

*   Data in wrong format
*   Wrong data


*   Duplicates



```python
import pandas as pd
dataset=pd.read_csv("Files/pandas_dataset.csv")
dataset
```

```python
#Data cleaning
dataset['Date']=dataset['Date'].str.replace("'","",regex=True)
dataset.head(4)
```

```python
#Return a dataset with no Empty cell
#df.dropna() will remove empty cells but with not change the dataframe
new_df=dataset.dropna()
new_df
```

```python
#another way of handling the null values is by replacing the null values with df.fillna()
df_new2=dataset.fillna(130)
df_new2
```

```python
df_new3 = dataset["Calories"].fillna("pun")
df_new3
```

```python
#Fill specific columns null value
df_new3=dataset
x=df_new3["Calories"].mean()
df_new3["Calories"]=df_new3["Calories"].fillna(x)
df_new3
```

```python
df_new4=dataset
df_new4["Date"]=dataset["Date"].fillna("2025/02/16")
df_new4
```

##Video Games Sales

```python
import pandas as pd
video_games_sales=pd.read_csv("Files/Video_Games_Sales_as_at_22_Dec_2016.csv")
video_games_sales
```

```python
df1=video_games_sales
a=df1["Critic_Score"].mean()
df1["Critic_Score"]=df1["Critic_Score"].fillna(a)
df1
```

```python
df2=video_games_sales
b=df2["Critic_Count"].mean()
df2["Critic_Count"]=df2["Critic_Count"].fillna(b)
df2
```

```python
df3=video_games_sales
c=df3["User_Score"].mean()
df3["User_Score"]=df3["User_Score"].fillna(c)
df3
```

```python
df4=video_games_sales
d=df4["User_Count"].mean()
df4["User_Count"]=df4["User_Count"].fillna(d)
df4
```

```python
df5=video_games_sales
e=df5["Developer"].mode()[0]
df5["Developer"]=df5["Developer"].fillna(e)
df5
```

```python
df6=video_games_sales
f=df6["Rating"].mode()[0]
df6["Rating"]=df6["Rating"].fillna(f)
df6
```
