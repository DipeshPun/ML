Data preprocessing is required tasks for cleaning the data and making it suitable for a machine learning model which also increases the accuracy and efficiency of a machine learning model.

It is built on top of another package name Numpy, which provides supports for multiple dimensional arrays. Preprocessing is the process of doing pre-analysis of data , in order to transform them into a standard and normalized from.

Preprocessing involves the following aspects:


*   missing values
*   data standardization
*   data normalization
*   data binning

```python
#Return a dataset with no Empty cell
#df.dropna() will remove empty cells but with not change the dataframe
new_df=df.dropna()
new_df
```
