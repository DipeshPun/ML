```python
import pandas as pd
import numpy as np
df=pd.read_excel("Files/Customer Call List.xlsx")
df
```

```python
#Make phone number in same format
df['Phone_Number']=df['Phone_Number'].astype(str).str.replace(r'\D','',regex=True)
df
```

```python
#Drop not usefull column
df.drop(columns=['Not_Useful_Column'],inplace=True, errors='ignore')
df
```

```python
#Phone number format in 3 digit(i.e 123-123-1234)
df['Phone_Number'] = df['Phone_Number'].str.replace(r'(\d{3})-?(\d{3})-?(\d{4}).*', r'\1-\2-\3', regex=True)
df
```

```python
#Drop duplicates
df_1 = df.drop_duplicates()
df_1 = df.drop_duplicates(subset=['Phone_Number'], keep='first')
df_1 = df.drop_duplicates(subset=['CustomerID'])
df_1
```

```python
#Last name
df_1['Last_Name'] = df_1['Last_Name'].str.replace(r'[^a-zA-Z ]', '', regex=True).str.strip()
df_1
```

```python
#Split the column Address and create three columns named: street_address, state, zip_code
df_1[['Street_Address', 'State', 'Zip_Code']] = df_1['Address'].str.split(',', expand=True)
df_1
```

```python
#For phone number fill Nan, N/A, n/a
df_1['Phone_Number'] = df_1['Phone_Number'].replace(['', ' ', 'NaN', 'N/a', 'n/a', 'NaN', None], np.nan)
df_1
```

```python
#In do not contact column
#replace yes to Y and No to N
df_1['Do_Not_Contact'] = df_1['Do_Not_Contact'].replace(['Yes', 'No'], ['Y', 'N'])
df_1
```

```python
#Entire Null values should be replaced with blank
df_1 = df_1.fillna('')
df_1
```

```python
#Remove rows if do not contact has the vale "Y"
df_1['Do_Not_Contact']=df_1['Do_Not_Contact'].replace('Y','')
df_1
```

```python
#Drop rows if phone number column is in empty string
df_1 = df_1[df_1['Phone_Number'] != '']
df_1
```
