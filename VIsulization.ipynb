## Why Choose Matplotlib and Seaborn?
Both libraries are among the most popular chices for data visualization in Python. Here's why:
Matplotlib:Known for its flexibility, Matplotlin allows detailed customization of visual elements, maing it perfect for tailored plots.
Seaborn: Built on thop of Matlaplib, Seaborn simplifies visualization tasks by offering hihg-level functions to create aesthetically pleasing statistical graphics.

###Visualization with Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Basics of Matplotlib
#Line plot
x=np.linspace(0, 10, 100)
y=np.sin(x)
plt.plot(x,y)
```

```python
x=np.linspace(0, 10, 100)
y=np.sin(x)
plt.plot(x,y,label="Sine wave")
plt.title("Line plot example")
plt.xlabel("x-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

```python
#Bar chart
categories=['A', 'B','C']
values=[10, 20 ,30]
plt.bar(categories, values, color=['Red', 'Green', 'Blue'])
ax=plt.gca()
ax.set_facecolor("Pink")
plt.title("Bar chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

###Visualization with Seaborn

```python
import seaborn as sns
```

```python
#Histogram and KDE pilot:
data=np.random.randn(1000)
sns.histplot(data, kde=True, color="Blue", edgecolor="Black")
plt.title("Histogram with KDE")
plt.show()
```

```python
#Heat Map
matrix=np.random.randn(5,5)
sns.heatmap(matrix, annot=True, cmap="plasma")
plt.title("Heat Map")
plt.show()
```

```python
#Pairplot example
sns.set_style("whitegrid")
df=sns.load_dataset("iris")
sns.pairplot(df, hue="species", markers=['o','s','D'])
plt.show()
```

```python
#Chess Board
import numpy as np
import matplotlib.pyplot as plt

# Create a checkerboard pattern (8x8 grid)
n = 8  # size of the chessboard
checkerboard = np.zeros((n, n))

# Fill the checkerboard with alternating colors (1s and 0s)
for i in range(n):
    for j in range(n):
        if (i + j) % 2 == 0:
            checkerboard[i, j] = 1  # Color 1 (e.g., white)
        else:
            checkerboard[i, j] = 0  # Color 2 (e.g., black)

# Create a heatmap with the chessboard pattern
plt.imshow(checkerboard, cmap='gray', interpolation='nearest')
plt.colorbar()  # Add a color bar if you want to display the scale

# Remove all borders and ticks
plt.axis('off')  # Hide the axes
plt.show()
```

```python
import pandas as pd
ice_cream=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Ice Cream Ratings.csv")
ice_cream
```

```python
#set date as index
ice_cream=ice_cream.set_index("Date")
ice_cream
```

```python
#Scatter plot using texture rating and overall rating
ice_cream.plot(kind='scatter', x='Texture Rating', y='Overall Rating', color='blue', title="Scatter Plot")
```

```python
#Hist Plot
import seaborn as sns
ice_cream=np.random.randn(1000)
sns.histplot(ice_cream, kde=True, color="Blue", edgecolor="Black")
plt.title("Histogram with KDE")
plt.show()
```

```python
# Creating the boxplot
ice_cream.boxplot(grid=False)
plt.title("Boxplot of Ratings")
plt.show()
```

```python
#Pie chart
numeric_columns = ["Flavor Rating", "Texture Rating", "Overall Rating"]
avg_ratings = ice_cream[numeric_columns].mean()
plt.figure(figsize=(6,6))
plt.pie(avg_ratings, labels=avg_ratings.index, autopct='%1.1f%%', startangle=140, colors=["skyblue", "lightcoral", "lightgreen"])
plt.title("Average Ratings Distribution")
plt.show()
```
